## 前言
通过excel来快速制图是很方便的，那么为什么要用matplotlib呢，在用python进行数据处理过程中，比如搭建了一个数据模型，有真实值和预测值，这个时候快速用matplotlib来比对真实值和预测值的匹配情况，就很合适。

因此学习matplotlib要轻，有需求的时候简单了解一下相关即可，而不是一定要什么图表制图需求都要想着看能不能用matplotlib做出来，除了excel，还有很多其他的工具，那个方便喜欢用那个就用那个，只是在某些情况下，就是随手写点matplotlib的制图代码就是最合适的。

## 如何引入matplotlib
下面的引用语句已经是惯例了。 

第二局是让图片直接在ipynb上显示，不需要调用 `plt.show()` 了。

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## 一个简单的绘制真实值和预测值的例子

```python
xdata = [1,2,3,4,5]
ydata = [1.1,2.1,2.9,3.9,5.1]
predict_ydata = [1,2,3,4,5]

plt.plot(xdata, ydata, color='blue')
plt.plot(xdata, predict_ydata, color='red', linestyle='dashed')
```
