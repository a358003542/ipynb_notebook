## 简单单隐藏层感知机

和basic里面的SimpleFNN网络基本相同，只是用pytorch的实现，为了对比也没有进行其他优化。

```python
from pywander.datasets import MnistDataset
training_data = MnistDataset(train=True)
test_data = MnistDataset(train=False)
```

```python
training_data.plot_image(2)
```

```python
test_data.plot_image(0)
```

```python
from torch.utils.data import DataLoader
batch_size = 16

train_dataloader = DataLoader(training_data, batch_size=batch_size)
test_dataloader = DataLoader(test_data, batch_size=batch_size)

for X, y in test_dataloader:
    print(f"Shape of X [N, C, H, W]: {X.shape}")
    print(f"Shape of y: {y.shape} {y.dtype}")
    break
```

```python
from pywander.models import load_model
model = load_model( 'mnist', 'simple_mlp.pkl')
model.plot_progress()
model.test_batch(test_dataloader)
```

```python
from pywander.models import load_model
_,ax = plt.subplots(figsize=(16, 8))
model = load_model( 'mnist', 'simple_mlp2.pkl')
model.plot_progress(ax=ax, alpha=0.1)
model.test_batch(test_dataloader)
```
